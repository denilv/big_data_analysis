```python
def sigmoid(a, derivative = False):
    arr = np.array(a)
    if (derivative):
        return sigmoid(arr) * (1 -  sigmoid(arr))
    else:
        return 1 / (1 + np.exp(-arr))
```

```python
def erase_pics():
    filelist = glob.glob("pics/*.jpg")
    for f in filelist:
        os.remove(f)
```

```python
def show_weights(W):
    result = create_big_pic(W)
    show(result, size = 8)
    
def save_pic(W, name = "weights.jpg"):
    pic = create_big_pic(W)
    scipy.misc.imsave("pics/" + name, pic)
    
def create_big_pic(W):
    w = np.array(W)
    w = w.T
    img_len = w.shape[1]
    result = np.zeros((280, 280))
    for i in xrange(10):
        for j in xrange(10):
            n = i * 10 + j
            img = w[n].reshape(28, 28)
            x = i * 28
            y = j * 28
            result[x : x + 28, y : y + 28] = img
#             print result[x : x + 28, y : y + 28].shape
    return result

def show(image, size = 3):
    fig = plt.figure(figsize = (size, size))
    ax = fig.add_subplot(1,1,1)
    imgplot = ax.imshow(image, cmap=mpl.cm.gray)
#     imgplot.set_interpolation('nearest')
    plt.axis('off')
    plt.show()

    
```

## Реализация ограниченной машины Больцмана

```python
class RBM():
    
    def __init__(self, n_visible = 28*28, n_hidden = 1000, visible_bias = None, \
                 hidden_bias = None, n_epochs = 1500, batch_size = 50, learning_rate = 0.01,\
                 regularization_type = None, regularization_rate = 0.01, momentum = 0.9, \
                 debug = False):
        
        if hidden_bias is None:
            hidden_bias = np.zeros(n_hidden)  # initialize h bias 0

        if visible_bias is None:
            visible_bias = np.zeros(n_visible)  # initialize v bias 0
        
        self.h_bias = hidden_bias
        self.v_bias = visible_bias
        self.n_visible = n_visible
        self.n_hidden = n_hidden
        self.n_epochs = n_epochs
        self.batch_size = batch_size
        self.lr = learning_rate
        self.reg_type = regularization_type
        self.reg_rate = regularization_rate
        self.momentum = momentum
        self.dw = 0.0
        self.da = 0.0
        self.db = 0.0
        self.scores = []
        self.weights = np.random.normal(loc=0.0, scale = 0.01, size=(n_visible, n_hidden))
        self.debug = debug
    
    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
        return self
    
    def get_params(self, deep=True):
        d = dict()
        d["n_visible"] = self.n_visible
        d["n_hidden"] = self.n_hidden
        d["n_epochs"] = self.n_epochs
        d["learning_rate"] = self.lr
        d["batch_size"] = self.batch_size
        return d
    
    def hid_step(self, vis):
        return sigmoid(np.dot(vis, self.weights) + self.h_bias)
    
    def vis_step(self, hid):
        return sigmoid(np.dot(hid, self.weights.T) + self.v_bias)
    
    def sample_hid(self, hid_probs):
        rand = np.random.uniform(size = hid_probs.shape)
        result = np.array([1.0 if (p > r) else 0.0 for p, r in zip(hid_probs, rand)])
        return result
        
    def sample_vis(self, vis_probs):
        rand = np.random.uniform(size = vis_probs.shape)
        result = np.array([1.0 if (p > r) else 0.0 for p, r in zip(vis_probs, rand)])
        return result
    
    def sample(self, probs):
        return (probs > np.random.uniform(size = probs.shape)).astype(float)
    
#     def regulate(self):
#         if self.reg_type is not None:
#             if self.reg_type.lower() == 'l1':
#                 return self.reg_rate * np.sign(self.weights)
#             if self.reg_type.lower() == 'l2':
#                 return self.reg_rate * self.weights
#         return 0
    
    def contr_div(self, batch):
        
        vis0 = batch
        hid_probs0 = self.hid_step(vis0)
        hid0 = self.sample(hid_probs0)
        vis_probs1 = self.vis_step(hid0)
        vis1 = self.sample(vis_probs1)
        hid_probs1 = self.hid_step(vis1)
        hid1 = self.sample(hid_probs1)
        
#         dw = (1.0 - self.momentum) * (np.dot(vis0.T, hid0) - np.dot(vis1.T, hid1)) / len(batch)
#         da = (1.0 - self.momentum) * (vis0 - vis1).mean(axis=0)
#         db = (1.0 - self.momentum) * (hid0 - hid1).mean(axis=0)
        
        dw = (np.dot(vis0.T, hid0) - np.dot(vis1.T, hid1)) / len(batch) #???
        da = (vis0 - vis1).mean(axis=0)
        db = (hid0 - hid1).mean(axis=0)
        
        #sparsity
#         gb = self.target_sparsity - h0.mean(axis=1)
        
        #calculate momentum
        mw = self.momentum * self.dw
        ma = self.momentum * self.da
        mb = self.momentum * self.db
        
        #finally dw, da, db
#         dw = self.lr * (dw + mw - self.reg_rate * self.weights)
#         da = self.lr * (da + ma)
#         db = self.lr * (db + mb)
        
        self.dw = dw
        self.da = da
        self.db = db
               
        dw = self.lr * (dw - self.reg_rate * self.weights) + mw
        da = self.lr * da + ma
        db = self.lr * db + mb
        
        self.weights += dw
        self.v_bias += da
        self.h_bias += db
        
        return vis1


    def fit(self, X, Y=None):
        train = X
        for it in range(self.n_epochs):
#             np.random.shuffle(train)
            sys.stdout.write('\r%s epoch of %s' % (it, self.n_epochs))
            sys.stdout.flush()
            if self.debug:
                if (it % 2 == 0):
                    if (it >= 100):
                        name =  str(it) + ".jpg"
                        save_pic(self.weights, name)
                    elif (it >= 10):
                        name = "0" + str(it) + ".jpg"
                        save_pic(self.weights, name)
                    else:
                        name = "00" + str(it) + ".jpg"
                        save_pic(self.weights, name)
            if it > 10:
                self.momentum = 0.9
            
            mses = []
            for i in xrange(len(X) / self.batch_size):
                batch = train[i * self.batch_size : (i + 1) * self.batch_size]
                self.contr_div(batch)
        sys.stdout.write('\r%s epoch of %s\n' % (self.n_epochs, self.n_epochs))
        if self.debug:
            show_weights(self.weights)
            save_pic(self.weights, "result.jpg")
        return self
    
    def transform(self, X, sample = False):
#         result = self.hid_step(self.sample(self.vis_step(self.sample(self.hid_step(X))))) 
#         if sample:
#             result = self.sample(result)
        return self.hid_step(X)
    
    def reconstruct(self, X):
        result = self.vis_step(self.sample(self.hid_step(X)))
        return result
    
    
```

```python
def count_mse(n = 100, rbm = None, imgs = np.array([])):
    all_mse = 0.0
    for i in xrange(n):
        img = np.copy(mntrainx[np.random.randint(5000)])
        half_img = np.append(img[0:500], [0.0 for i in xrange(284)], axis = 0)
        reconstructed_img = rbm.reconstruct(half_img)
        curr_mse =  np.sum((img - reconstructed_img)**2) / len(img)
        all_mse += curr_mse
    return all_mse / n
```
