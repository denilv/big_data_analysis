```python
def show_weights(W):
    result = create_big_pic(W)
    show(result, size = 8)
    
def save_pic(W, name = "weights.jpg"):
    pic = create_big_pic(W)
    scipy.misc.imsave("pics/" + name, pic)
    
def create_big_pic(W):
    w = np.array(W)
    w = w.T
    n_imgs = int(np.sqrt(w.shape[0]))
    img_len = w.shape[1]
    result = np.zeros((280, 280))
    for i in xrange(n_imgs):
        for j in xrange(n_imgs):
            n = i * n_imgs + j
            img = w[n].reshape(28, 28)
            x = i * 28
            y = j * 28
            result[x : x + 28, y : y + 28] = img
#             print result[x : x + 28, y : y + 28].shape
    return result

def show(image, size = 3):
    fig = plt.figure(figsize = (size, size))
    ax = fig.add_subplot(1,1,1)
    imgplot = ax.imshow(image, plt.cm.gray)
#     imgplot.set_interpolation('nearest')
    plt.axis('off')
    plt.show()

def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.001)
    return tf.Variable(initial)

def bias_variable(shape):
    initial = tf.constant(0.0, shape=shape)
    return tf.Variable(initial)
```

```python
class VAE():
    def __init__(self, network_architecture = [784, 400, 20, 400, 784], func=tf.nn.softplus,
                 learning_rate=0.001, batch_size=100, n_epochs = 25):
        self.network_architecture = network_architecture
        self.func = func
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.n_epochs = n_epochs

        self.x = tf.placeholder(tf.float32, [None, network_architecture[0]])

        self._create_network()
        self._create_loss_optimizer()

        init = tf.global_variables_initializer()

        self.sess = tf.InteractiveSession()
        self.sess.run(init)

    def _create_network(self):
        self.W, self.b = self._initialize_weights()

        self.z_mean, self.z_log_sigma = self._enc_network()

        eps = tf.random_normal([self.batch_size, self.network_architecture[2]],  mean=0.0, stddev=1.0, \
                               dtype=tf.float32)
        # z = mu + sigma*epsilon
        #self.z = tf.add(self.z_mean, tf.mul(tf.sqrt(tf.exp(self.z_log_sigma)), eps))
        self.z = tf.add(self.z_mean, tf.mul(tf.exp(self.z_log_sigma), eps))

        self.x_reconstr_mean = self._dec_network()

    def _initialize_weights(self):
        def weight_variable(shape):
            initial = tf.truncated_normal(shape, stddev=0.001)
            return tf.Variable(initial)

        def bias_variable(shape):
            initial = tf.constant(0.0, shape=shape)
            return tf.Variable(initial)

        W = dict()
        b = dict()
        W['inp_enc'] = weight_variable([self.network_architecture[0], self.network_architecture[1]])
        b['enc'] = bias_variable([self.network_architecture[1]])

        W['enc_z_mean'] = weight_variable([self.network_architecture[1], self.network_architecture[2]])
        W['enc_z_log_sigma'] = weight_variable([self.network_architecture[1], self.network_architecture[2]])

        b['z_mean'] = bias_variable([self.network_architecture[2]])
        b['z_log_sigma'] = bias_variable([self.network_architecture[2]])

        W['z_dec'] = weight_variable([self.network_architecture[2], self.network_architecture[3]])
        b['dec'] = bias_variable([self.network_architecture[3]])

        W['dec_out_mean'] = weight_variable([self.network_architecture[3], self.network_architecture[4]])
        W['dec_out_log_sigma'] = weight_variable([self.network_architecture[3], self.network_architecture[4]])
        b['out_mean'] = bias_variable([self.network_architecture[4]])
        b['out_log_sigma'] = bias_variable([self.network_architecture[4]])
        return W, b

    def _enc_network(self):
        enc_layer = self.func(tf.add(tf.matmul(self.x, self.W['inp_enc']), self.b['enc']))
        z_mean = tf.add(tf.matmul(enc_layer, self.W['enc_z_mean']), self.b['z_mean'])
        z_log_sigma = tf.add(tf.matmul(enc_layer, self.W['enc_z_log_sigma']), self.b['z_log_sigma'])
        return z_mean, z_log_sigma

    def _dec_network(self):
        dec_layer = self.func(tf.add(tf.matmul(self.z, self.W['z_dec']), self.b['dec']))
        x_reconstr_mean = tf.nn.sigmoid(tf.add(tf.matmul(dec_layer, self.W['dec_out_mean']), self.b['out_mean']))
        return x_reconstr_mean

    def _create_loss_optimizer(self):
        reconstr_loss = -tf.reduce_sum(self.x * tf.log(1e-10 + self.x_reconstr_mean) + (1-self.x) * \
                                       tf.log(1e-10 + 1 - self.x_reconstr_mean),\
                                       axis=1)
        latent_loss = -0.5 * tf.reduce_sum(1 + 2.0 * self.z_log_sigma - tf.square(self.z_mean) \
                                           - tf.exp(2.0 * self.z_log_sigma), \
                                           axis=1)
        self.cost = tf.reduce_mean(reconstr_loss + latent_loss)
        self.optimizer = tf.train.AdamOptimizer(learning_rate=self.learning_rate).minimize(self.cost)
        
    def fit(self, X):
        X_sh = shuffle(X)
        for epoch in range(self.n_epochs):
            avg_cost = 0.
            n_batches = int(len(X) / self.batch_size)
            for i in xrange(n_batches):
                batch = X_sh[i * self.batch_size: (i+1)*self.batch_size]
                opt, cost = self.sess.run((self.optimizer, self.cost), feed_dict={self.x: batch})
                avg_cost += cost / len(X) * self.batch_size

            if epoch % 5 == 0:
                print "Epoch:", '%04d' % (epoch), "cost=", "{:.9f}".format(avg_cost)
        return avg_cost

    def transform(self, X):
        return self.sess.run(self.z_mean, feed_dict={self.x: X})

    def generate(self, z_mu=None):
        if z_mu is None:
            z_mu = np.random.normal(size=[self.batch_size, self.network_architecture[2]])
        rec_one = self.func(tf.add(tf.matmul(self.z, self.W['z_dec']), self.b['dec']))
        return self.sess.run(self.x_reconstr_mean,
                             feed_dict={self.z: z_mu})

    def reconstruct(self, X):
        return self.sess.run(self.x_reconstr_mean,
                             feed_dict={self.x: X})

```
